```python
class Fraction:
    """
    A class to represent mathematical fractions with numerator and denominator.
    Supports basic arithmetic operations and fraction simplification.
    """

    def __init__(self, n, d):
        """
        Initialize a fraction with numerator n and denominator d.
        Args:
            n: numerator (integer)
            d: denominator (integer, should not be zero)
        """
        # Validate denominator
        if d == 0:
            raise ValueError("Denominator cannot be zero")

        self.num = n
        self.den = d
        self._simplify()  # Automatically simplify fractions on creation

    def _gcd(self, a, b):
        """
        Calculate the Greatest Common Divisor using Euclidean algorithm.
        Private helper method for fraction simplification.
        """
        while b:
            a, b = b, a % b
        return abs(a)

    def _simplify(self):
        """Simplify the fraction by dividing numerator and denominator by their GCD."""
        gcd = self._gcd(self.num, self.den)
        self.num //= gcd
        self.den //= gcd

        # Ensure denominator is positive
        if self.den < 0:
            self.num = -self.num
            self.den = -self.den

    def __str__(self):
        """Return string representation of the fraction."""
        return f"{self.num}/{self.den}"

    def __repr__(self):
        """Return detailed string representation for debugging."""
        return f"Fraction({self.num}, {self.den})"

    def __add__(self, other):
        """Addition operator: self + other"""
        if not isinstance(other, Fraction):
            # Allow addition with integers
            other = Fraction(other, 1)

        temp_num = self.num * other.den + other.num * self.den
        temp_den = self.den * other.den
        return Fraction(temp_num, temp_den)

    def __sub__(self, other):
        """Subtraction operator: self - other"""
        if not isinstance(other, Fraction):
            other = Fraction(other, 1)

        temp_num = self.num * other.den - other.num * self.den
        temp_den = self.den * other.den
        return Fraction(temp_num, temp_den)

    def __mul__(self, other):
        """Multiplication operator: self * other"""
        if not isinstance(other, Fraction):
            other = Fraction(other, 1)

        temp_num = self.num * other.num
        temp_den = self.den * other.den
        return Fraction(temp_num, temp_den)

    def __truediv__(self, other):
        """Division operator: self / other"""
        if not isinstance(other, Fraction):
            other = Fraction(other, 1)

        # Multiply by reciprocal
        temp_num = self.num * other.den
        temp_den = self.den * other.num
        return Fraction(temp_num, temp_den)

    def __eq__(self, other):
        """Equality operator: self == other"""
        if not isinstance(other, Fraction):
            other = Fraction(other, 1)

        # Cross multiply to compare fractions
        return self.num * other.den == other.num * self.den

    def __lt__(self, other):
        """Less than operator: self < other"""
        if not isinstance(other, Fraction):
            other = Fraction(other, 1)

        return self.num * other.den < other.num * self.den

    def __le__(self, other):
        """Less than or equal operator: self <= other"""
        return self < other or self == other

    def __gt__(self, other):
        """Greater than operator: self > other"""
        if not isinstance(other, Fraction):
            other = Fraction(other, 1)

        return self.num * other.den > other.num * self.den

    def __ge__(self, other):
        """Greater than or equal operator: self >= other"""
        return self > other or self == other

    def __neg__(self):
        """Unary negation operator: -self"""
        return Fraction(-self.num, self.den)

    def __abs__(self):
        """Absolute value operator: abs(self)"""
        return Fraction(abs(self.num), self.den)

    def __pow__(self, exponent):
        """Power operator: self ** exponent (only integer exponents)"""
        if not isinstance(exponent, int):
            raise TypeError("Exponent must be integer")

        if exponent >= 0:
            return Fraction(self.num ** exponent, self.den ** exponent)
        else:
            # For negative exponents, return reciprocal
            return Fraction(self.den ** abs(exponent), self.num ** abs(exponent))

    def get_value(self):
        """Return the decimal value of the fraction."""
        return self.num / self.den

    def reciprocal(self):
        """Return the reciprocal of the fraction."""
        return Fraction(self.den, self.num)

    def is_proper(self):
        """Check if the fraction is proper (numerator < denominator)."""
        return abs(self.num) < abs(self.den)
```

```python
# Example usage and testing
# Create fractions
f1 = Fraction(3, 4)
f2 = Fraction(1, 2)
f3 = Fraction(2, 4)  # Will be automatically simplified to 1/2

print(f"f1 = {f1}")  # 3/4
print(f"f2 = {f2}")  # 1/2
print(f"f3 = {f3}")  # 1/2 (automatically simplified)
```

```python
# Arithmetic operations
print(f"\n{f1} + {f2} = {f1 + f2}")  # 5/4
print(f"{f1} - {f2} = {f1 - f2}")  # 1/4
print(f"{f1} * {f2} = {f1 * f2}")  # 3/8
print(f"{f1} / {f2} = {f1 / f2}")  # 3/2
```

```python
# Comparison operations
print(f"\n{f1} == {f2}: {f1 == f2}")  # False
print(f"{f2} == {f3}: {f2 == f3}")    # True
print(f"{f1} > {f2}: {f1 > f2}")      # True
```

```python
# Unary operations
print(f"\nNegative of {f1}: {-f1}")        # -3/4
print(f"Absolute of {-f1}: {abs(-f1)}")    # 3/4
```

```python
# Powers
print(f"\n{f1} ** 2 = {f1 ** 2}")          # 9/16
print(f"{f1} ** -1 = {f1 ** -1}")          # 4/3
```

```python
# Additional methods
print(f"\nDecimal value of {f1}: {f1.get_value():.2f}")  # 0.75
print(f"Reciprocal of {f1}: {f1.reciprocal()}")          # 4/3
print(f"Is {f1} proper? {f1.is_proper()}")               # True
```
