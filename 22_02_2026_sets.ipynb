## Frozen set

- Immutable set.
- Created via `frozenset()`.
- No modifications post-creation.
- Ideal for hashable items: **strings**, **numbers**.

```python
# create frozenset
fs1 = frozenset([1, 2, 3])
fs2 = frozenset([3, 4, 5])

print(fs1,type(fs1))
print(fs2,type(fs2))
```

```python
# create frozenset
fs1 = frozenset([1, 2, 3])
fs2 = frozenset([3, 4, 5])
fs1 | fs2
```

```python
# 5. Frozen sets (immutable sets)
frozen = frozenset([1, 2, 3, 4, 5])

print(f"Frozen set: {frozen}")
print(f"Type: {type(frozen)}")
```

```python
# Frozen set operations (read-only)
frozen2 = frozenset([4, 5, 6, 7, 8])
print(f"Union of frozen sets: {frozen | frozen2}")
```

```python
frozen.add(6) # This would raise AttributeError
```

```python
# what works and what does not
# Works       ---> all read functions
# Doesn't Work ---> write operations
```

```python
# When to use
# 2D sets
fs = frozenset([1, 2, frozenset([3, 4])])
fs
```

## Set Comprehension

```python
# examples
{i**2 for i in range(1, 11)}
```

```python
{i**2 for i in range(1, 11) if i>5}
```

```python
# Filtered set comprehension
# Set of the Square of the even numbers in the range of 1 to 10
even_squares = {x**2 for x in range(1, 11) if x % 2 == 0}
print(f"Even squares: {even_squares}")
```

```python
# Set of the Square root of the odd numbers in the range of 9 to 20
even_sqrt = {x**(1/2) for x in range(9, 21) if x % 2 != 0}
print(f"Even square root: {even_sqrt}")
```
