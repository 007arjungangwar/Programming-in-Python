```python
num=145
original_num=num
sum_of_fact_last_digit=0
while num>0:
  last_digit=num%10
  num//=10
  last_digit_fact=1
  for i in range(1,last_digit+1):
    last_digit_fact*=i
  sum_of_fact_last_digit+=last_digit_fact

if sum_of_fact_last_digit==original_num:
  print("Strong Number")
else:
  print("Not a Strong Number")
```
