```python
str1="what are you doing tomorrow"

# Using a loop and count() method
for char in str1:
  if str1.count(char) == 1:
    print(f"The first non-repeating character is: {char}")
    break

# If prefer a while loop :
i=0
while i < len(str1):
  if str1.count(str1[i]) == 1:
    print(f"The first non-repeating character is: {str1[i]}")
    break
  i+=1
```

```python
# Count Uppercase, Lowercase, Digits, Special Characters

str1="ArjunGangwariitpkd@gmail.com"


upper = lower = digit = special = dot= 0

for ch in range(len(str1)):
  if "A"<=str1[ch]<="Z":
    upper+=1
  elif "a"<=str1[ch]<="z":
    lower+=1

  elif "0"<=str1[ch]<="9":
    digit+=1
  else:
    special+=1

print(upper,lower,digit,special)
```

```python
# Check if Two Strings are Anagrams

word1="listen"
word2="silent"
angram=True
word=None
for i in range(len(word1)):
  word=word1[i]
  if word1.count(word)==word2.count(word):
    angram=True
    continue
  else:
    angram=False

if angram:
  print("Anagram")
else:
  print("Not Anagram")
```
