```python
# Check if Two Strings are Anagrams

word1="listen"
word2="silent"
angram=True
word=None
for i in range(len(word1)):
  word=word1[i]
  if word1.count(word)==word2.count(word):
    angram=True
    continue
  else:
    angram=False

if angram:
  print("Anagram")
else:
  print("Not Anagram")
```

```python
# zip() combines multiple iterables
names = ["Alice", "Bob", "Charlie"]
scores = [85, 92, 78]
ages = [20, 22, 21]
# Basic zip usage
zipped = zip(names, scores, ages)
print("Zipped result:", list(zipped))
# Practical use: Calculate weighted scores
weights = [0.3, 0.4, 0.3]
grades = ["A", "B", "C"]
for name, score, weight, grade in zip(names, scores, weights, grades):
  weighted = score * weight
  print(f"{name}: Score={score}, Weight={weight}, Weighted={weighted:.1f}, Grade={grade}")
```

```python
str = "madam"
is_palindrome = 1 # True
length = 0

# Find length
for ch in str:
  length = length + 1

# Check palindrome
i = 0
j = length- 1
while i < j:
  if str[i] != str[j]:
    is_palindrome = 0 # False
  break
  i = i + 1
  j = j- 1

if is_palindrome == 1:
  print(str, "is a palindrome")
else:
  print(str, "is not a palindrome")
```
